# Plot 
training accuracy log

```python
import h5py
import matplotlib.pyplot as plt
file_name = './results/uschad_ProtoHAR_0.001_0.0_1.0_3u_14tu_32b_5_20k2p_0.h5'
# Open the .h5 file
with h5py.File(file_name, 'r') as hf:
    rs_glob_acc = hf['rs_glob_acc'][:]
    rs_f1 = hf['rs_f1'][:]
    rs_auc = hf['rs_auc'][:]
    rs_train_acc = hf['rs_train_acc'][:]
    rs_train_loss = hf['rs_train_loss'][:]
```

```python
# Plot global accuracy
plt.figure()
plt.plot(rs_glob_acc, label='Global Accuracy')
plt.xlabel('Iterations')
plt.ylabel('Accuracy')
plt.title('Global Accuracy over Iterations')
plt.legend()
out_file = file_name.replace('.h5', '_glob_acc.png')
out_file = out_file.replace('./results/', './plotresult/')
plt.savefig(out_file)
plt.show()
```

```python

# Plot F1 score
plt.figure()
plt.plot(rs_f1, label='F1 Score')
plt.xlabel('Iterations')
plt.ylabel('F1 Score')
plt.title('F1 Score over Iterations')
plt.legend()
out_file = file_name.replace('.h5', '_f1.png')
out_file = out_file.replace('./results/', './plotresult/')
plt.savefig(out_file)
plt.show()
```

```python

# Plot AUC
plt.figure()
plt.plot(rs_auc, label='AUC')
plt.xlabel('Iterations')
plt.ylabel('AUC')
plt.title('AUC over Iterations')
plt.legend()
out_file = file_name.replace('.h5', '_auc.png')
out_file = out_file.replace('./results/', './plotresult/')
plt.savefig(out_file)
plt.show()
```

```python

# Plot training accuracy
plt.figure()
plt.plot(rs_train_acc, label='Training Accuracy')
plt.xlabel('Iterations')
plt.ylabel('Accuracy')
plt.title('Training Accuracy over Iterations')
plt.legend()
out_file = file_name.replace('.h5', '_train_acc.png')
out_file = out_file.replace('./results/', './plotresult/')
plt.savefig(out_file)
plt.show()
```

```python

# Plot training loss
plt.figure()
plt.plot(rs_train_loss, label='Training Loss')
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.title('Training Loss over Iterations')
plt.legend()
out_file = file_name.replace('.h5', '_train_loss.png')
out_file = out_file.replace('./results/', './plotresult/')
plt.savefig(out_file)
plt.show()
```
